```python
#Dropping columns having null percentage > 30
def nullcolumns(df,percent):
    delcols = (df.isnull().sum()/len(df))
    delcols = list(delcols[delcols.values >= percent].index)
    df.drop(labels=delcols, axis=1, inplace=True)
    return df
```

```python
app_data = nullcolumns(app_data,percent=0.30)
```
